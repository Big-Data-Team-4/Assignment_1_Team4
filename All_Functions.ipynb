```python
import pandas as pd
from jiwer import wer, cer
from sklearn.metrics import precision_score, recall_score, f1_score

def evaluate_parsing(gt_text_path, pred_text_path, gt_csv_path, pred_csv_path):
    results = {}
    
    # --- 1. TEXT EVALUATION (WER & CER) ---
    with open(gt_text_path, "r") as f:
        gt_text = f.read().strip()
    with open(pred_text_path, "r") as f:
        pred_text = f.read().strip()
    
    results["WER"] = wer(gt_text, pred_text)
    results["CER"] = cer(gt_text, pred_text)

    # --- 2. TABLE EVALUATION (Precision & Recall at cell level) ---
    gt_df = pd.read_csv(gt_csv_path).fillna("").astype(str)
    pred_df = pd.read_csv(pred_csv_path).fillna("").astype(str)

    # Align shapes (truncate to min rows/cols)
    min_rows = min(gt_df.shape[0], pred_df.shape[0])
    min_cols = min(gt_df.shape[1], pred_df.shape[1])

    gt_vals = gt_df.iloc[:min_rows, :min_cols].values.flatten()
    pred_vals = pred_df.iloc[:min_rows, :min_cols].values.flatten()

    # Ground truth = actual cell content
    # Prediction = does it match or not?
    y_true = [1 if g.strip() != "" else 0 for g in gt_vals]   # 1 = should be a valid cell
    y_pred = [1 if g.strip() == p.strip() else 0 for g, p in zip(gt_vals, pred_vals)]

    # Metrics
    results["Table_Precision"] = precision_score(y_true, y_pred, zero_division=0)
    results["Table_Recall"] = recall_score(y_true, y_pred, zero_division=0)
    results["Table_F1"] = f1_score(y_true, y_pred, zero_division=0)

    return results


# ðŸ”¹ Example usage:
results = evaluate_parsing(
    gt_text_path="/Users/anushaprakash/Desktop/Bigdata/WER/groundtruth_text.txt",
    pred_text_path="/Users/anushaprakash/Desktop/Bigdata/src/data/parsed/page_texts/Intel10k_2025/page_010.txt",
    gt_csv_path="/Users/anushaprakash/Desktop/Bigdata/WER/financial_indices_with_dollar.csv",
    pred_csv_path="/Users/anushaprakash/Desktop/Bigdata/WER/table_page_69_pymupdf_page_69_block_5_stream_enhanced.csv"
)

print("Evaluation Results:")
for k, v in results.items():
    print(f"{k}: {v:.3f}")

```

```python
#!/usr/bin/env python3
"""
Simple evaluation for single files - Project LANTERN Part 9
Evaluates one text file and one CSV file against ground truth
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from pathlib import Path
from datetime import datetime
from typing import Dict, Any
from difflib import SequenceMatcher

def compute_text_wer_cer(ground_truth_file: str, extracted_file: str) -> Dict[str, float]:
    """
    Compute WER and CER for single text files
    """
    def levenshtein_distance(s1: str, s2: str) -> int:
        if len(s1) < len(s2):
            return levenshtein_distance(s2, s1)
        
        if len(s2) == 0:
            return len(s1)
        
        previous_row = list(range(len(s2) + 1))
        for i, c1 in enumerate(s1):
            current_row = [i + 1]
            for j, c2 in enumerate(s2):
                insertions = previous_row[j + 1] + 1
                deletions = current_row[j] + 1
                substitutions = previous_row[j] + (c1 != c2)
                current_row.append(min(insertions, deletions, substitutions))
            previous_row = current_row
        
        return previous_row[-1]
    
    # Read files
    with open(ground_truth_file, 'r', encoding='utf-8') as f:
        gt_text = f.read()
    
    with open(extracted_file, 'r', encoding='utf-8') as f:
        ext_text = f.read()
    
    # Clean and normalize
    gt_clean = re.sub(r'\s+', ' ', gt_text.strip().lower())
    ext_clean = re.sub(r'\s+', ' ', ext_text.strip().lower())
    
    # Word Error Rate
    gt_words = gt_clean.split()
    ext_words = ext_clean.split()
    
    word_distance = levenshtein_distance(' '.join(gt_words), ' '.join(ext_words))
    wer = word_distance / len(gt_words) if gt_words else 0
    
    # Character Error Rate
    char_distance = levenshtein_distance(gt_clean, ext_clean)
    cer = char_distance / len(gt_clean) if gt_clean else 0
    
    similarity_ratio = SequenceMatcher(None, gt_clean, ext_clean).ratio()
    
    return {
        'WER': float(wer),
        'CER': float(cer),
        'similarity_ratio': float(similarity_ratio),
        'ground_truth_length': len(gt_clean),
        'extracted_length': len(ext_clean),
        'word_count_gt': len(gt_words),
        'word_count_extracted': len(ext_words)
    }

def compute_table_precision_recall(ground_truth_csv: str, extracted_csv: str) -> Dict[str, float]:
    """
    Compute table precision and recall for single CSV files
    """
    try:
        # Load CSVs
        gt_df = pd.read_csv(ground_truth_csv)
        ext_df = pd.read_csv(extracted_csv)
        
        print(f"Ground truth shape: {gt_df.shape}")
        print(f"Extracted shape: {ext_df.shape}")
        print(f"Ground truth columns: {list(gt_df.columns)}")
        print(f"Extracted columns: {list(ext_df.columns)}")
        
        # Normalize data
        gt_df = gt_df.astype(str).applymap(lambda x: str(x).strip().lower())
        ext_df = ext_df.astype(str).applymap(lambda x: str(x).strip().lower())
        
        # Take minimum dimensions for comparison
        min_rows = min(len(gt_df), len(ext_df))
        min_cols = min(len(gt_df.columns), len(ext_df.columns))
        
        print(f"Comparing dimensions: {min_rows} rows Ã— {min_cols} columns")
        
        if min_rows == 0 or min_cols == 0:
            return {
                'table_precision': 0.0,
                'table_recall': 0.0,
                'table_f1': 0.0,
                'exact_matches': 0,
                'total_cells': 0,
                'error': 'No data to compare'
            }
        
        gt_aligned = gt_df.iloc[:min_rows, :min_cols]
        ext_aligned = ext_df.iloc[:min_rows, :min_cols]
        
        # Cell-by-cell comparison
        total_cells = min_rows * min_cols
        exact_matches = 0
        partial_matches = 0
        
        for i in range(min_rows):
            for j in range(min_cols):
                gt_cell = str(gt_aligned.iloc[i, j]).strip()
                ext_cell = str(ext_aligned.iloc[i, j]).strip()
                
                if gt_cell == ext_cell:
                    exact_matches += 1
                elif is_partial_match(gt_cell, ext_cell):
                    partial_matches += 1
        
        # Calculate metrics
        precision = exact_matches / total_cells if total_cells > 0 else 0
        recall = exact_matches / total_cells if total_cells > 0 else 0
        f1_score = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        
        return {
            'table_precision': float(precision),
            'table_recall': float(recall),
            'table_f1': float(f1_score),
            'exact_matches': exact_matches,
            'partial_matches': partial_matches,
            'total_cells': total_cells,
            'match_percentage': (exact_matches / total_cells * 100) if total_cells > 0 else 0
        }
        
    except Exception as e:
        return {
            'table_precision': 0.0,
            'table_recall': 0.0,
            'table_f1': 0.0,
            'error': str(e)
        }

def is_partial_match(gt_cell: str, ext_cell: str, threshold: float = 0.8) -> bool:
    """Check if cells are partial matches"""
    if not gt_cell or not ext_cell:
        return False
    
    # Normalize financial text
    gt_norm = normalize_financial_text(gt_cell)
    ext_norm = normalize_financial_text(ext_cell)
    
    similarity = SequenceMatcher(None, gt_norm, ext_norm).ratio()
    return similarity >= threshold

def normalize_financial_text(text: str) -> str:
    """Normalize financial text for comparison"""
    normalized = text.replace(',', '').replace('$', '').replace('(', '-').replace(')', '')
    normalized = re.sub(r'\s+', ' ', normalized.strip())
    return normalized

def create_visualization(text_metrics: Dict, table_metrics: Dict, save_path: str = "evaluation_results.png"):
    """Create visualization of evaluation results"""
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle('Project LANTERN - Single File Evaluation Results', fontsize=14)
    
    # Text metrics
    text_scores = [1 - text_metrics['WER'], 1 - text_metrics['CER'], text_metrics['similarity_ratio']]
    text_labels = ['1 - WER', '1 - CER', 'Similarity']
    
    ax1.bar(text_labels, text_scores, color=['blue', 'green', 'orange'])
    ax1.set_title('Text Extraction Quality')
    ax1.set_ylabel('Score (higher is better)')
    ax1.set_ylim(0, 1)
    
    # Add values on bars
    for i, v in enumerate(text_scores):
        ax1.text(i, v + 0.02, f'{v:.3f}', ha='center', va='bottom')
    
    # Table metrics
    if 'error' not in table_metrics:
        table_scores = [table_metrics['table_precision'], table_metrics['table_recall'], table_metrics['table_f1']]
        table_labels = ['Precision', 'Recall', 'F1-Score']
        
        ax2.bar(table_labels, table_scores, color=['red', 'purple', 'brown'])
        ax2.set_title('Table Extraction Quality')
        ax2.set_ylabel('Score')
        ax2.set_ylim(0, 1)
        
        for i, v in enumerate(table_scores):
            ax2.text(i, v + 0.02, f'{v:.3f}', ha='center', va='bottom')
    else:
        ax2.text(0.5, 0.5, f"Table Error:\n{table_metrics['error']}", 
                ha='center', va='center', transform=ax2.transAxes)
        ax2.set_title('Table Extraction - Error')
    
    # Text length comparison
    lengths = [text_metrics['word_count_gt'], text_metrics['word_count_extracted']]
    length_labels = ['Ground Truth', 'Extracted']
    
    ax3.bar(length_labels, lengths, color=['lightblue', 'lightcoral'])
    ax3.set_title('Word Count Comparison')
    ax3.set_ylabel('Word Count')
    
    for i, v in enumerate(lengths):
        ax3.text(i, v + max(lengths) * 0.02, str(v), ha='center', va='bottom')
    
    # Table cell matches (if available)
    if 'error' not in table_metrics and table_metrics['total_cells'] > 0:
        matches = [table_metrics['exact_matches'], 
                  table_metrics['partial_matches'],
                  table_metrics['total_cells'] - table_metrics['exact_matches'] - table_metrics['partial_matches']]
        match_labels = ['Exact', 'Partial', 'No Match']
        colors = ['green', 'yellow', 'red']
        
        ax4.pie(matches, labels=match_labels, colors=colors, autopct='%1.1f%%', startangle=90)
        ax4.set_title('Table Cell Match Distribution')
    else:
        ax4.text(0.5, 0.5, "No table data\nto display", ha='center', va='center')
        ax4.set_title('Table Cell Matches - No Data')
    
    plt.tight_layout()
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
    
    print(f"Visualization saved to: {save_path}")

def run_single_file_evaluation(
    ground_truth_text: str,
    ground_truth_table: str,
    extracted_text: str,
    extracted_table: str,
    output_dir: str = "evaluation_results"
) -> Dict[str, Any]:
    """
    Run evaluation on single files
    """
    
    # Create output directory
    Path(output_dir).mkdir(exist_ok=True)
    
    print("="*60)
    print("PROJECT LANTERN - SINGLE FILE EVALUATION")
    print("="*60)
    
    # Evaluate text
    print("\n1. Evaluating Text Extraction...")
    print(f"Ground Truth: {ground_truth_text}")
    print(f"Extracted: {extracted_text}")
    
    text_metrics = compute_text_wer_cer(ground_truth_text, extracted_text)
    
    print(f"âœ“ WER: {text_metrics['WER']:.4f}")
    print(f"âœ“ CER: {text_metrics['CER']:.4f}")
    print(f"âœ“ Similarity: {text_metrics['similarity_ratio']:.4f}")
    
    # Evaluate table
    print("\n2. Evaluating Table Extraction...")
    print(f"Ground Truth: {ground_truth_table}")
    print(f"Extracted: {extracted_table}")
    
    table_metrics = compute_table_precision_recall(ground_truth_table, extracted_table)
    
    if 'error' not in table_metrics:
        print(f"âœ“ Precision: {table_metrics['table_precision']:.4f}")
        print(f"âœ“ Recall: {table_metrics['table_recall']:.4f}")
        print(f"âœ“ F1-Score: {table_metrics['table_f1']:.4f}")
        print(f"âœ“ Exact Matches: {table_metrics['exact_matches']}/{table_metrics['total_cells']} cells ({table_metrics['match_percentage']:.1f}%)")
    else:
        print(f"âœ— Table Error: {table_metrics['error']}")
    
    # Create visualization
    print("\n3. Creating Visualization...")
    viz_path = Path(output_dir) / "evaluation_visualization.png"
    create_visualization(text_metrics, table_metrics, str(viz_path))
    
    # Generate summary report
    print("\n4. Generating Report...")
    results = {
        'timestamp': datetime.now().isoformat(),
        'files': {
            'ground_truth_text': ground_truth_text,
            'ground_truth_table': ground_truth_table,
            'extracted_text': extracted_text,
            'extracted_table': extracted_table
        },
        'text_metrics': text_metrics,
        'table_metrics': table_metrics,
        'overall_quality': {
            'text_quality': 1 - text_metrics['WER'],
            'table_quality': table_metrics.get('table_f1', 0),
            'combined_score': (1 - text_metrics['WER']) * 0.6 + table_metrics.get('table_f1', 0) * 0.4
        }
    }
    
    # Save results
    import json
    results_path = Path(output_dir) / "evaluation_results.json"
    with open(results_path, 'w') as f:
        json.dump(results, f, indent=2)
    
    print(f"âœ“ Results saved to: {results_path}")
    
    # Print summary
    print("\n" + "="*60)
    print("EVALUATION SUMMARY")
    print("="*60)
    print(f"Text Quality (1-WER): {results['overall_quality']['text_quality']:.4f}")
    print(f"Table Quality (F1):   {results['overall_quality']['table_quality']:.4f}")
    print(f"Combined Score:       {results['overall_quality']['combined_score']:.4f}")
    
    return results

def main():
    """Main function with your specific file paths"""
    
    # Your file paths
    ground_truth_text = "/Users/anushaprakash/Desktop/Bigdata/WER/groundtruth_text.txt"
    ground_truth_table = "/Users/anushaprakash/Desktop/Bigdata/WER/groundtruth_table.csv"
    extracted_text = "/Users/anushaprakash/Desktop/Bigdata/src/data/parsed/pdfparser_output/page_texts/intel10k_2024/page_010.txt"
    extracted_table = "/Users/anushaprakash/Desktop/Bigdata/src/data/parsed/camelot_output/intel10k_2024/tables_csv/lattice_basic_table_11_page_48.csv"
    
    # Run evaluation
    results = run_single_file_evaluation(
        ground_truth_text=ground_truth_text,
        ground_truth_table=ground_truth_table,
        extracted_text=extracted_text,
        extracted_table=extracted_table,
        output_dir="evaluation_results"
    )
    
    return results

if __name__ == "__main__":
    main()
```
